```python
import geopandas as gpd
import folium
from folium.plugins import FastMarkerCluster
rides = rideshare[rideshare['destination']=='Financial District']
map = folium.Map(location=[rides["latitude"].mean(), rides["longitude"].mean()],zoom_start = 10)
map = map.add_child(FastMarkerCluster(rides[['latitude', 'longitude']].values.tolist()))
map
```
